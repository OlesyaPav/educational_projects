Видим, что столбец "рейтинг" заполнен разными типами оценок -строками в виде дробный чисел и в виде процентов. Как такое могло произойти?


Создадим функкцию, которая из строки с числом и знаком процента делает целое число между 0 и 10, в соответствии с оценками фильма на Кинопоиске.

```python
def f(s):
    s = str(s)
    try:
        if ('%' in s):
            return int(s.replace("%",""))/10
        else:
            return float(s)
    except:
        return 0

```

```python
data['ratings1'] = data['ratings'].apply(f) #создадим новый столбец рейтинга с числами от 0 до 10.
```

```python
 del data['ratings']#Удалим старый столбец "Рейтинг".
```

```python
data.sample(5)
```

```python
data.info()
```

*Вывод: Добавили новый столбец "Рейтинг" с вещественным типом. По какой-то, видимо, техничсекой причине в колонках вопреки оценкам в Кинопоиске находились оценки в процентах. Вид оценок приведен к одному - десятибальному.*
