#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

### Создание признаков

Создадим функцию, которая формирует признаки для временного ряда.

```python
def make_features(data, max_lag):
    data['hour'] = data.index.hour
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(24, max_lag*24+1, 24):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
    
    data['rolling_mean'] = data['num_orders'].shift().rolling(24).mean()
```

```python
make_features(data, 5)

data.head(50)
```

Увеличение столбцов отставания увеличивает качество модели (проверено опытным путем на кросс-валидации).

*Итак, созданная функция сформировала признаки из часа, месяца, даты, дня недели трех столбцов отстающих значений по историческим данным и столбца со скользящим средним по историческим данным.*

### Подготовка данных для обучения

```python
train, test = train_test_split(data, shuffle = False, test_size=0.1)
```

```python
display(train.tail(5))
test.head(5)
```

```python
print('Размер тренировочной выборки',train.shape)
print('Размер тестовой выборки',test.shape)
print((train.shape[0]+test.shape[0]) == data.shape[0])
```

```python
train = train.dropna()
train.shape
```

```python
X_train = train.drop('num_orders',axis=1)
y_train = train.num_orders
X_test = test.drop('num_orders',axis=1)
y_test = test.num_orders

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
X_train.to_csv('X_train.csv')
y_train.to_csv('y_train.csv')
X_test.to_csv('X_test.csv')
y_test.to_csv('y_test.csv')
```

Разделили данные на тренировочные и тестовые (10%). Сверили конец тренировоных данных и начало тестовых. Проверили размеры данных. Отделили входные и целевой признаки, проверили их размеры.

```python
num_columns = ['lag_24', 'lag_48', 'lag_72', 'lag_96', 'lag_120','rolling_mean']

ohe_columns = [] #остальные столбцы отразили как категориальные
for i in X_train.columns:
    if (i not in num_columns): 
        ohe_columns.append(i)
print('Категориальные столбцы: ohe_columns =',ohe_columns)
```

```python
ohe_pipe = Pipeline(
    [('ohe', OneHotEncoder(drop='first', sparse_output=False,
                          handle_unknown='error',
                          categories=[list(range(0, 25)), 
                                      list(range(1, 13)), 
                                      list(range(1, 32)),
                                      list(range(0, 7))
                                     ]))
     ])

data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe,ohe_columns)],
     remainder='passthrough'
)
```

### Ridge

```python
pipe = Pipeline([
    ('preprocessor', data_preprocessor),
    ('ridge', Ridge())
])

params = {
    'ridge__alpha': [0.1, 1, 10, 100],
    'ridge__solver': ['auto', 'svd', 'cholesky']
}

tscv = TimeSeriesSplit()
grid = GridSearchCV(pipe,
                    params,
                    cv=tscv,
                    scoring='neg_root_mean_squared_error',
                    n_jobs=-1)
grid.fit(X_train, y_train)
print("Лучшие параметры:", grid.best_params_)
print("Лучший RMSE:", -round(grid.best_score_,2))
```

### Random Forest

```python
pipe = Pipeline([
    ('preprocessor', data_preprocessor),
    ('regressor', RandomForestRegressor(random_state=RANDOM_STATE))
])

params = {
    'regressor__n_estimators': [50, 100, 200],
    'regressor__max_depth': [None, 5, 10],
    'regressor__min_samples_split': [2, 5, 10],
    'regressor__max_features': ['sqrt', 'log2']
}

tscv = TimeSeriesSplit()
grid_search_forest = GridSearchCV(
    pipe,
    params,
    cv=tscv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1)

grid_search_forest.fit(X_train, y_train)

print("Лучшие параметры:", grid_search_forest.best_params_)
print("Лучший RMSE:", -round(grid_search_forest.best_score_,2))
```

###  CatBoost

```python
model = CatBoostRegressor(
    cat_features=ohe_columns,
    early_stopping_rounds=50,
    random_state=RANDOM_STATE,
    verbose=False
)

params = {
     'iterations': [500, 1000],  
     'learning_rate': [0.01, 0.05, 0.1],
     'depth': [4, 6, 8], 
     'l2_leaf_reg': [1, 3, 5] 
     }
tscv = TimeSeriesSplit()

grid_search_cat = GridSearchCV(
    model,
    params,
    cv=tscv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1)

grid_search_cat.fit(X_train, y_train)

print("Лучшие параметры:", grid_search_cat.best_params_)
print("Лучший RMSE:", round(-grid_search_cat.best_score_,2))
```
